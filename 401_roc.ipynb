# Rate of Change indicator ( ROC)

### About ROC

- ROC is also part of the oscillators group ( like MACD, RSI etc)

- Momentum Indicator : Measures the strength of the momentum based on rate of change in price.
    
- Is plotted against a Zero line
    - Rising ROC or positive values above Zero indicate upward momentum or upward buying pressure
    - Falling ROC or negative values below Zero indicate downward momentum or selling pressure

- returns = (current price - previous price) / previous price
- ROC = (current price - n period ago price) / n period ago price


- current price = 110
- n day previous price = 100 

- ROC(n) = return = 10 % or 10 per cent  10 / 100 = 0.10

## Features
### 1. Zero-line crossovers can be used to signal trend changes




### 2. Overbought and oversold levels - Look for extreme ROC values which resulted in price reversals in the past




### 3. Also used as a divergence indicator , indicating upcoming trend change

- Divergence occurs when the price of a stock or another asset moves in one direction while its ROC moves in the opposite direction



### Limitations

1. ROC is prone to whipsaws, especially around the zero line. 
1. Calculation gives equal weight to the most recent price and the price from n periods ago
1. Generally not used for trading purposes, but mainly to alert a trend change.

### Takeaway 
- Like all technical indicators, the Rate-of-Change oscillator should be used in conjunction with other aspects of technical analysis.
